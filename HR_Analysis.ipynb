```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd. read_csv('HR_Analytics.csv')
```

```python
df.isnull().sum()
```

```python
print(f"Dataset Shape: {df.shape}")
```

```python
df.describe()
```

```python
sns.set_theme(style="whitegrid")
#1. Attrition count
plt.figure(figsize=(6, 2))
sns.countplot(x='Attrition', data=df, palette='pastel' , hue="Attrition")
plt.title('Attrition Count', fontsize=14)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Attrition', fontsize=12)
plt.show()
```

```python
# 2. Attrition by Gender
plt.figure(figsize=(6,3))
sns.countplot(x='Gender', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition by Gender', fontsize=14)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Gender', fontsize=12)
plt.legend(title="Attrition")
plt.show()
```

```python
# 3. Monthly Income Distribution by Attrition
plt.figure(figsize=(6,3))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='coolwarm',hue = "Attrition")
plt.title('Monthly Income Distribution by Attrition', fontsize=14)
plt.ylabel('Monthly Income', fontsize=12)
plt.xlabel('Attrition', fontsize=12)
plt.show()
```

```python
# 5. Attrition by Business Travel
plt.figure(figsize=(6,3))
sns.countplot(x='BusinessTravel', hue='Attrition', data=df, palette='muted')
plt.title('Attrition by Business Travel', fontsize=14)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Business Travel Frequency', fontsize=12)
plt.legend(title="Attrition")
plt.show()
```

```python
# 6. Age Distribution
plt.figure(figsize=(6,3))
sns.histplot(df['Age'], kde=True, color='blue', bins=20)
plt.title('Age Distribution of Employees', fontsize=14)
plt.ylabel('Frequency', fontsize=12)
plt.xlabel('Age', fontsize=12)
plt.show()
```

```python
# 7. Job Satisfaction vs. Attrition
plt.figure(figsize=(6,3))
sns.boxplot(x='Attrition', y='JobSatisfaction', data=df, palette='coolwarm',hue = 'Attrition')
plt.title('Job Satisfaction vs. Attrition', fontsize=14)
plt.ylabel('Job Satisfaction', fontsize=12)
plt.xlabel('Attrition', fontsize=12)
plt.show()
```

HR analytics reveals key insights: Attrition rates are influenced by job satisfaction, travel frequency, and income levels. Younger employees show higher attrition, while those with frequent travel are more likely to leave. Optimizing satisfaction can reduce turnover.
